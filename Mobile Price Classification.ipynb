```python
X_test_full = X_test_full.drop_duplicates()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
```

```python
X
```

```python
X_test = sc.fit_transform(X_test_full)
```

```python
X_test
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from xgboost import XGBClassifier , XGBRFClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report as cr , confusion_matrix as cm , accuracy_score as ac
```

```python
logreg_without_overfitting = LogisticRegression(C = 0.01)
logreg_with_overfitting = LogisticRegression()
dtree = DecisionTreeClassifier()
rf = RandomForestClassifier(n_estimators=500)
xgc = XGBClassifier()
xgbRFc = XGBRFClassifier()
```

```python
def model(model):
    model.fit(X_train,y_train)
    ypred = model.predict(X_test)
    print(cr(y_test,ypred))
    print("Accuracy Score is :",ac(y_test,ypred)*100)
```

```python
model(dtree)
```

```python
model(rf)
```

```python
model(logreg_with_overfitting)
```

```python
model(logreg_without_overfitting)
```

```python
model(xgc)
```

```python
model(xgbRFc)
```
